Berdasarkan data diatas kita dapat melihat sekilas bahwa terdapat informasi age, anaemia, creatine_phosphokinase, dan lain-lain adalah feature, dan DEATH_EVENT adalah label/target. 

```python
#outlier detection
def detect_outliers(df, n, features):
    outlier_indices = []
    '''
    Mendeteksi outlier dari daftar fitur yang diberikan. 
    Ini mengembalikan daftar indeks menurut pengamatan yang 
    mengandung lebih dari n outlier menurut Tukey Method
    '''
    
    #iterasi feature(kolom)
    for col in features:
        q1 = np.percentile(df[col], 25)
        q3 = np.percentile(df[col], 75)
        iqr = q3 - q1
        
        #outlier step
        outlier_step = 1.5 * iqr
        
        #menentukan daftar indeks outlier untuk kolom feature
        outlier_list_col = df[(df[col] < q1 - outlier_step) | (df[col] > q3 + outlier_step)].index
        
        #menambahkan indeks outlier yang ditemukan untuk col ke daftar indeks outlier
        outlier_indices.extend(outlier_list_col)
        
    #pilih observation yang mengandung lebih dari 2 outlier
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, j in outlier_indices.items() if j > n)
    
    return multiple_outliers

#mendeteksi outlier dari feature numeric
outliers_detected = detect_outliers(df, 2, ["age","anaemia","creatinine_phosphokinase","platelets", "serum_creatinine", "serum_sodium", "time"])
```

Disini kami menggunakan Tukey Method. Tukey Method adalah prosedur perbandingan berganda satu langkah (single-step multiple) dan uji statistik. Ini dapat digunakan untuk menemukan means(rata-rata) yang signifikan berbeda satu sama lain. Kami mendeteksi outlier dari feature numeric (age, anaemia, platelets, dan lain-lain). Lalu, kami menganggap outlier sebagai baris yang memiliki setidaknya 2 nilai outlier

## Cross Validate Model

Saya akan menggunakan 4 klasifikasi model yang populer dan mengevaluasi model-model tersebut dengan menggunakan mean accuracy kfold cross validation
1. Random Forest
2. AdaBoost
3. Gradient Boosting
4. Hist Gradient Boosting
5. XGBoost

```python
#membuat function untuk evaluate model
def evaluate_model(models):
    #cross val model with kfold
    kfold = StratifiedKFold(n_splits = 10)
    
    result = []
    for model in models:
        result.append(cross_val_score(model, x_train, y_train, scoring='accuracy', cv=kfold, n_jobs=4))
        
    cv_means = []
    cv_std = []
    
    for cv_result in result:
        cv_means.append(cv_result.mean())
        cv_std.append(cv_result.std())
        
    df_new = pd.DataFrame({
        'CrossValMeans': cv_means,
        'CrossValError': cv_std,
        'Models' : [
            'Random Forest Classifier',
            'AdaBoost Classifier',
            'Gradient Boosting Classifier',
            'Hist Gradient Boosting Classifier',
            'XGBoost Classifier'
        ]
    })
    
    #generate chart
    bar = sns.barplot(x='CrossValMeans', y='Models', data=df_new, orient='h')
    bar.set_xlabel('Mean Accuracy')
    bar.set_title('Cross Validation Scores')
    
    return df_new
```

```python
#modelling
random_state = 30
models = [
     RandomForestClassifier(random_state = random_state),
     AdaBoostClassifier(DecisionTreeClassifier(random_state = random_state), random_state = random_state, learning_rate = 0.2),
     GradientBoostingClassifier(random_state = random_state),
     HistGradientBoostingClassifier(random_state=random_state),
     XGBClassifier(random_state=random_state)
]

evaluate_model(models)
```

Berdasarkan dari cross validation scores di atas, nilai cross val mean yang tinggi berdasarkan urutan yaitu Random Forest, Gradient Boosting, XGBoost, Hist Gradient Boosting, dan AdaBoost.

## Learning Curves

Dengan mempelajari kemiringan (learning curve) dapat membantu kami untuk menemukan model yang overfitting yang mana dapat mempengaruhi akurasi model

```python
def show_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)):
    """Generate plot sederhana dari test dan train learning curve"""
    plt.figure()
    plt.title(title)
    
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")
    plt.legend(loc="best")
```

```python
show_learning_curve(gs_rf.best_estimator_, 'Random Forest Learning Curve', x_train, y_train, cv=kfold)
show_learning_curve(gs_ab.best_estimator_,'AdaBoost Learning Curve', x_train, y_train, cv=kfold)
show_learning_curve(gs_gb.best_estimator_,'Gradient boosting Learning Curve', x_train, y_train, cv=kfold)
show_learning_curve(gs_hg.best_estimator_,'Hist Gradient Boosting Learning Curve', x_train, y_train, cv=kfold)
show_learning_curve(gs_xg.best_estimator_,'XGBoost Learning Curve', x_train, y_train, cv=kfold)
```

Berdasarkan Learning Curve di atas, dapat dilihat bahwa Random Forest, Hist Gradient Boosting, dan XGBoost overfitting. AdaBoost dan Gradient Boosting terlihat lebih baik dalam menggeneralisasikan prediksi. Selanjutnya ensemble model untuk model voting

# Ensemble Model: Voting

```python
#menggabungkan model
voting = VotingClassifier(estimators=[
    ('gs_rf', gs_rf.best_estimator_),
    ('gs_ab', gs_ab.best_estimator_),
    ('gs_gb',gs_gb.best_estimator_),
    ('gs_hg',gs_hg.best_estimator_),
    ('gs_xg',gs_xg.best_estimator_)], voting='soft', n_jobs = -1)

voting = voting.fit(x_train, y_train)
```

Pada variabel voting berisikan Voting Classifier. Voting Classifier adalah estimator pembelajaran mesin yang melatih berbagai model dasar atau estimator dan memprediksi berdasarkan penggabungan temuan dari masing-masing estimator atau model dasar. 
Pada ensembel model voting ini, kami menggabungkan model yang telah ditentukan sebelumnya berdasarkan cross validation score, yaitu Random Forest, AdaBoost, Gradient Boosting, Hist Gradient Boosting, dan XGBoost.

```python
#prediksi
from sklearn import metrics

y_pred  = voting.predict(x_test)
print(metrics.classification_report(y_test, y_pred))
```

Dari classification report diatas, dapat kita ambil kesimpulan bahwa terdapat kelas Label 0 dan 1 yang mana memiliki nilai precision, recall, f1-score, dan support.
Preicision adalah kemampuan pengklasifikasi untuk tidak melabeli instance positif yang sebenarnya negatif. Akurasi prediksi positif.
Recall adalah kemampuan classifier untuk menemukan semua contoh positif. Fraksi positif yang diidentifikasi dengan benar.
F1-score adalah rata-rata harmonik tertimbang precision dan recall sehingga skor terbaik adalah 1,0 dan yang terburuk adalah 0,0. Skor F1 lebih rendah dari ukuran akurasi karena mereka menanamkan precision dan recall ke dalam perhitungannya. Sebagai patokan, rata-rata tertimbang F1 harus digunakan untuk membandingkan model pengklasifikasi, bukan akurasi global.
Support adalah jumlah kemunculan sebenarnya dari kelas dalam kumpulan data yang ditentukan. Dukungan yang tidak seimbang dalam data pelatihan dapat menunjukkan kelemahan struktural dalam skor yang dilaporkan dari pengklasifikasi dan dapat menunjukkan perlunya stratified sampling atau rebalancing. Dukungan tidak berubah antar model melainkan mendiagnosis proses evaluasi.

- f1-score = 0.93, angka yang menunjukkan mendekati 1.0 adalah model yang lebih baik. Karena 0.93 mendekati 1.0 maka dapat dikatakan model nya sudah bagus.
