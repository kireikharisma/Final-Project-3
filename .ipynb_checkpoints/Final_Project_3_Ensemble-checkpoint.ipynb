Berdasarkan data diatas kita dapat melihat sekilas bahwa terdapat informasi age, anaemia, creatine_phosphokinase, dan lain-lain adalah feature, dan DEATH_EVENT adalah label/target. 

```python
#outlier detection
def detect_outliers(df, n, features):
    outlier_indices = []
    '''
    Mendeteksi outlier dari daftar fitur yang diberikan. 
    Ini mengembalikan daftar indeks menurut pengamatan yang 
    mengandung lebih dari n outlier menurut Tukey Method
    '''
    
    #iterasi feature(kolom)
    for col in features:
        q1 = np.percentile(df[col], 25)
        q3 = np.percentile(df[col], 75)
        iqr = q3 - q1
        
        #outlier step
        outlier_step = 1.5 * iqr
        
        #menentukan daftar indeks outlier untuk kolom feature
        outlier_list_col = df[(df[col] < q1 - outlier_step) | (df[col] > q3 + outlier_step)].index
        
        #menambahkan indeks outlier yang ditemukan untuk col ke daftar indeks outlier
        outlier_indices.extend(outlier_list_col)
        
    #pilih observation yang mengandung lebih dari 2 outlier
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, j in outlier_indices.items() if j > n)
    
    return multiple_outliers

#mendeteksi outlier dari feature numeric
outliers_detected = detect_outliers(df, 2, ["age","anaemia","creatinine_phosphokinase","platelets", "serum_creatinine", "serum_sodium", "time"])
```

Disini kami menggunakan Tukey Method. Tukey Method adalah prosedur perbandingan berganda satu langkah (single-step multiple) dan uji statistik. Ini dapat digunakan untuk menemukan means(rata-rata) yang signifikan berbeda satu sama lain. Kami mendeteksi outlier dari feature numeric (age, anaemia, platelets, dan lain-lain). Lalu, kami menganggap outlier sebagai baris yang memiliki setidaknya 2 nilai outlier

# Feature Engineering

```python
#get_dummies categorical features
df = pd.get_dummies(df, columns=['anaemia'], prefix='anae')
df = pd.get_dummies(df, columns=['diabetes'], prefix='dia')
df = pd.get_dummies(df, columns=['high_blood_pressure'], prefix='hbp')
df = pd.get_dummies(df, columns=['sex'], prefix='sex')
df = pd.get_dummies(df, columns=['smoking'], prefix='smoking')
```

# Modelling

## Data Splitting

```python
#train and test dataset
features = df.drop(['DEATH_EVENT'], axis=1)
label = df['DEATH_EVENT']

x_train, x_test, y_train, y_test = train_test_split(features, label, test_size=0.3, random_state=7)
```

## Cross Validate Model

Saya akan menggunakan 8 klasifikasi model yang populer dan mengevaluasi model-model tersebut dengan menggunakan mean accuracy kfold cross validation
1. Logistic Regression
2. KNN
3. SVM
4. Naive Bayes
5. Decision Tree
6. Random Forest
7. AdaBoost
8. Gradient Boosting

```python
#membuat function untuk evaluate model
def evaluate_model(models):
    #cross val model with kfold
    kfold = StratifiedKFold(n_splits = 10)
    
    result = []
    for model in models:
        result.append(cross_val_score(model, x_train, y_train, scoring='accuracy', cv=kfold, n_jobs=4))
        
    cv_means = []
    cv_std = []
    
    for cv_result in result:
        cv_means.append(cv_result.mean())
        cv_std.append(cv_result.std())
        
    df_new = pd.DataFrame({
        'CrossValMeans': cv_means,
        'CrossValError': cv_std,
        'Models' : [
            'Logistic Regression',
            'KNeighbors Classifier',
            'SVM',
            'Naive Bayes',
            'Decision Tree Classifier',
            'Random Forest Classifier',
            'AdaBoost Classifier',
            'Gradient Boosting Classifier'
        ]
    })
    
    #generate chart
    bar = sns.barplot(x='CrossValMeans', y='Models', data=df_new, orient='h')
    bar.set_xlabel('Mean Accuracy')
    bar.set_title('Cross Validation Scores')
    
    return df_new
```

```python
#modelling
random_state = 30
models = [
     LogisticRegression(random_state = random_state),
     KNeighborsClassifier(),
     SVC(random_state = random_state),
     GaussianNB(),
     DecisionTreeClassifier(random_state = random_state),
     RandomForestClassifier(random_state = random_state),
     AdaBoostClassifier(DecisionTreeClassifier(random_state = random_state), random_state = random_state, learning_rate = 0.2),
     GradientBoostingClassifier(random_state = random_state),
]

evaluate_model(models)
```

Berdasarkan dari cross validation scores di atas, nilai cross val mean yang tinggi akan digunakan untuk ensemble modelling. Jadi kami memutuskan untuk mengambil Logistic Regression, Naive Bayes, Random Forest Classifier, dan Gradient Boosting Classifier.

## Hyperparameter Tuning: untuk model yang dipilih

Penyetelan hyperparameter merupakan masalah dalam memilih satu set hyperparameter optimal untuk mempelajari algoritma. Hyperparameter adalah parameter yang nilainya digunakan untuk mengontrol proses pembelajaran.

```python
kfold = StratifiedKFold(n_splits=10)
```

```python
#random forest
rf = RandomForestClassifier()
rf_params = {
    'bootstrap': [False], 'max_depth': [None], 'max_features': [1, 5, 10], 
    'min_samples_leaf': [1, 4, 7], 'min_samples_split': [4, 8, 10], 
    'n_estimators': [100, 250], "criterion": ["gini"]
}

gs_rf = GridSearchCV(rf, param_grid=rf_params, cv=kfold, n_jobs=5, scoring='accuracy', verbose=1)
gs_rf.fit(x_test, y_test)

gs_rf.best_score_
```

```python
#logistic regression
lr = LogisticRegression()
lr_params = {
    'penalty' : ['l1', 'l2'],
    'C' : np.logspace(-4, 4, 20),
    'solver' : ['liblinear']
}

gs_lr = GridSearchCV(lr, param_grid=lr_params, cv=kfold, scoring='accuracy', verbose=1, n_jobs= 3)
gs_lr.fit(x_test, y_test)

gs_lr.best_score_
```

```python
#gradient boosting
gb = GradientBoostingClassifier()
gb_params = {
    'n_estimators' : [100,300],
    'learning_rate': [0.1, 0.05, 0.01],
    'min_samples_leaf': [4, 8, 10],
    'loss' : ["deviance"],
    'max_depth': [2, 8],
    'max_features': [0.3, 0.1] 
}

gs_gb = GridSearchCV(gb, param_grid=gb_params, cv=kfold, scoring='accuracy', n_jobs=3, verbose=1)
gs_gb.fit(x_test, y_test)

gs_gb.best_score_
```

```python
#Naive bayes
nb = GaussianNB()
nb_params = {
    'var_smoothing': np.logspace(0,-9, num=100)
}

gs_nb = GridSearchCV(nb, param_grid=nb_params, cv=kfold, scoring='accuracy', n_jobs=3, verbose=1)
gs_nb.fit(x_test, y_test)

gs_nb.best_score_
```

## Learning Curves

Dengan mempelajari kemiringan (learning curve) dapat membantu kami untuk menemukan model yang overfitting yang manadapat mempengaruhi akurasi model

```python
def show_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)):
    """Generate plot sederhana dari test dan train learning curve"""
    plt.figure()
    plt.title(title)
    
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")
    plt.legend(loc="best")
```

```python
show_learning_curve(gs_lr.best_estimator_, 'Logistic Regression Learning Curve', x_train, y_train, cv=kfold)
show_learning_curve(gs_rf.best_estimator_,'Random Forest Learning Curve', x_train, y_train, cv=kfold)
show_learning_curve(gs_gb.best_estimator_,'Gradient boosting Learning Curve', x_train, y_train, cv=kfold)
show_learning_curve(gs_nb.best_estimator_,'Naive Bayes Learning Curve', x_train, y_train, cv=kfold)
```

Berdasarkan Learning Curve di atas, dapat dilihat bahwa Random Forest dan Gradient Boosting overfitting. Logistic Regression dan Naive Bayes terlihat lebih baik dalam menggeneralisasikan prediksi. Selanjutnya ensemble model untuk model voting

## Ensemble Model: Voting

```python
#menggabungkan model
voting = VotingClassifier(estimators=[
    ('gs_lr', gs_lr.best_estimator_),
    ('gs_rf', gs_rf.best_estimator_),
    ('gs_gb',gs_gb.best_estimator_),
    ('gs_nb',gs_nb.best_estimator_)], voting='soft', n_jobs = 5)

voting = voting.fit(x_train, y_train)
```

```python
#prediksi
y_pred  = voting.predict(x_test)
accuracy = np.round(accuracy_score(y_test, y_pred) * 100, 3)
precision = np.round(precision_score(y_test, y_pred) * 100, 3)

print('Test dataset accuracy score: ', accuracy)
print('Test dataset precision score: ', precision)
```

Pada variabel voting berisikan Voting Classifier. Voting Classifier adalah estimator pembelajaran mesin yang melatih berbagai model dasar atau estimator dan memprediksi berdasarkan penggabungan temuan dari masing-masing estimator atau model dasar. 
Pada ensembel model voting ini, kami menggabungkan model yang telah ditentukan sebelumnya berdasarkan cross validation score, yaitu Logisctic Regression, Random Forest, Gradient Boosting, dan Naive Bayes

Setelah digabungkan, langkah selanjutnya yaitu memprediksi nilai. Dapat kita lihat bahwa hasil dari Test dataset accuracy score nya 87.778, sedangkan Test dataset precision score nya 85.0
